```python
#-*- encoding:utf-8 -*-
import time
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup
import csv

whole_source = ""
for i in range(21, 59):
    url='https://search.shopping.naver.com/search/category.nhn?pagingIndex='+str(i)+'&pagingSize=80&productSet=model&viewType=list&sort=rel&cat_id=50000193&frm=NVSHMDL'
    
    response = requests.get(url)
    
    whole_source = whole_source + response.text
    
bsObject = BeautifulSoup(whole_source, "html.parser")

# html = urlopen('https://search.shopping.naver.com/search/category.nhn?pagingIndex=1&pagingSize=80&productSet=model&viewType=list&sort=rel&cat_id=50000438&frm=NVSHCAT')
# bsObject = BeautifulSoup(html, "html.parser")


shopping_lists =[]
for shopping in bsObject.find_all('div',{"class":"tit"}):
    link = shopping.select('a')[0].get('href')
    shopping_lists.append(link)

for index, shopping_list in enumerate(shopping_lists,1):
    htmls = urlopen(shopping_list)
    soup = BeautifulSoup(htmls,"html.parser")

    title = soup.find('meta',{'property':'og:title'}).get('content')
    toner_title = title.replace(": 네이버쇼핑", "")
    
    brand_1 = soup.find('div',{'class':'info_inner'})
    brand = brand_1.find('em').string
    
    try:
        lotion_star = soup.find('div',{'class':'gpa'}).string
    except: 
        None
        
    try:
        e_type = soup.find('div',{'class':'info_inner list'})
        explain_1 = e_type.find('span').string


    except:
        None

    explain_2 = explain_1.replace("모든피부용","")
    explain_3 = explain_2.split(":")
    lotion_explain = explain_3[1]

    try:
        skin_type = soup.find('tr',{'class':'last'})
        a_type = skin_type.find('td').string.strip()    
    except:
        None
    lotion_type = a_type.replace("50ml(g)","")


    process = soup.find_all('span',{'class':'s_nowrap'})

    category_list = []
    for thing in process :
        category_list.append(thing.a.text)
        lotion_category = category_list[-1]

    try:     
        order_a = soup.find('a',{'data-tab-name':'review'})
        lotion_order = order_a.find('em').string
    except:
        
        None

    print(index, toner_title, brand, lotion_star, lotion_explain, lotion_type, lotion_category, lotion_order)
    time.sleep(0.5)

# #     data_header =['ranking','n_product','n_rating','explain','n_type','n_category','n_order']
    data = [[index, toner_title, brand, lotion_star, lotion_explain, lotion_type, lotion_category, lotion_order]]

    csvfile = open('C:\class\Mask.csv', 'a',-1, newline='', encoding='utf8') 
    csvwriter = csv.writer(csvfile)


    for row in data:    
        csvwriter.writerow(row)        

    csvfile.close()
```
